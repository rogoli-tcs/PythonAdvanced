```python
### Linked List implementation in Python3

class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None

class SLinkedList:
    def __init__(self):
        self.headval = None
        
    def listprint(self):
        currentNode = self.headval
        printlist = []
        while currentNode != None:
            printlist.append(currentNode.dataval)
            currentNode = currentNode.nextval
        print('->'.join(printlist))
        
    def insertatstart(self,newval):
        #print("Inserting Value %s at start"%newval)
        newnode = Node(newval)
        newnode.nextval = self.headval
        self.headval = newnode
        print(self.listprint())
        
    def insertatend(self,newval):
        #print("Inserting Value %s at end"%newval)
        newnode = Node(newval)
        
        if(self.headval==None):
            self.headval = newnode
            return
        lastnode = self.headval
        while(lastnode.nextval):
            lastnode = lastnode.nextval
        lastnode.nextval = newnode
        print(self.listprint())
        
    def insertatmiddle(self,middlenode,newval):
        #print("Inserting Value %s at middle"%newval)
        newnode = Node(newval)
        
        if(middlenode is None):
            print("Middle Node is not mentioned")
            return False
        
        newnode.nextval = middlenode.nextval
        middlenode.nextval = newnode
        print(self.listprint())
        
    def removenode(self,curval):
        #print("Remove %s from list"%curval)
        curnode = self.headval
        prenode = None
        searchflag = False
        while(curnode):
            if(curnode.dataval == curval):
                #print(curnode.dataval)
                searchflag = True
                break
            else:
                #print(curnode.dataval)
                prenode = curnode
                curnode = curnode.nextval
        
        if not searchflag:
            print("%s Value NOT Found!"%curval)
            return
        
        if prenode:
            prenode.nextval = curnode.nextval
        else:
            self.headval = curnode.nextval
        print(self.listprint())

        
list1 = SLinkedList()
list1.headval = Node("Mon")
e2 = Node("Tue")
e3 = Node("Wed")

list1.headval.nextval = e2
e2.nextval = e3

list1.listprint()
list1.insertatstart("Sun")
list1.insertatend("Fri")
list1.insertatmiddle(e3,"Thu")
list1.insertatend("Sat")
list1.removenode("Sat")
list1.removenode("Wed")
list1.removenode("Sun")
list1.removenode("Sun1")
list1.removenode("Fri")
```

```python
### Double Linked List

class Node:
    def __init__(self,data):
        self.preval = None
        self.dataval = data
        self.nextval = None
        
    def nodePrint(self):
        if(self.preval):
            preval = self.preval.dataval
        else:
            preval = 'None'
        if(self.nextval):
            nextval = self.nextval.dataval
        else:
            nextval = 'None'
        print("{} <- {} -> {}".format(preval,self.dataval,nextval))

class DLL:
    def __init__(self):
        self.headval = None
        
    def listprint(self):
        curnode = self.headval
        items = []
        while(curnode):
            items.append(curnode.dataval)
            #curnode.nodePrint()
            last = curnode
            curnode = curnode.nextval
        print(">> DLL : "+' -> '.join(items))
        
        items = []
        while(last):
            items.append(last.dataval)
            last = last.preval
            
        print("<< DLL : "+' <- '.join(items)+"\n")
        
    def insertatbegin(self,newval):
        newnode = Node(newval)
        self.headval.preval = newnode
        newnode.nextval = self.headval
        self.headval = newnode
        
        self.listprint()
        
    def append(self,newval):
        newnode = Node(newval)
        curnode = self.headval
        while(curnode):
            last = curnode
            curnode = curnode.nextval
            
        if(self.headval==None):
            self.headval = newnode
        else:
            last.nextval = newnode
            newnode.preval = last
        self.listprint()
        
    def remove(self,val):
        curnode = self.headval
        searchflag = False
        
        while(curnode):
            last = curnode.preval
            if(curnode.dataval == val):
                searchflag = True
                break
            curnode = curnode.nextval
        
        if(not searchflag):
            print("\n'%s' Value not found in DLL\n"%val)
            return
            
        if(curnode.preval == None):
            self.headval = curnode.nextval
            self.headval.preval = None
        else:
            last.nextval = curnode.nextval
            if(curnode.nextval):
                curnode.nextval.preval = last
            
        self.listprint()
    
list1 = DLL()
list1.append("Tue")
list1.insertatbegin("Mon")
list1.append("Wed")
list1.append("Thu")
## value not found
list1.remove("Sat")

list1.insertatbegin("Sun")
list1.insertatbegin("Sat")
list1.remove("Sat")
list1.append("Fri")
list1.append("Sat")
list1.remove("Wed")
list1.remove("Sat")
```

```python
### Binary Tree

class Node():
    def __init__(self,data):
        self.left = None
        self.right = None
        self.data = data
        
    def insert(self,data):
        if(self.data):
            if(data<self.data):
                if(self.left):
                    self.left.insert(data)
                else:
                    self.left = Node(data)
            elif(data>self.data):
                if(self.right):
                    self.right.insert(data)
                else:
                    self.right = Node(data)
        else:
            self.data = data
    
    def findval(self, lookup):
        if(lookup<self.data):
            if(self.left):
                return self.left.findval(lookup)
            else:
                return False
        elif(lookup>self.data):
            if(self.right):
                return self.right.findval(lookup)
            else:
                return False
        else:
            return True
        
    def printTree(self):
        if(self.left):
            self.left.printTree()
        print(self.data)
        if(self.right):
            self.right.printTree()
            
bt = Node(12)
bt.insert(3)
bt.insert(17)
bt.insert(20)
bt.insert(8)
bt.printTree()

print("------")
bt2 = Node("the")
bt2.insert("quick")
bt2.insert("brown")
bt2.insert("fox")
bt2.insert("jumps")
bt2.insert("over")
bt2.insert("lazy")
bt2.insert("dog")
bt2.printTree()
print('-------')
print(bt2.findval("apple"))
print(bt2.findval("quick"))
```

```python
## Heaps in Python
import heapq

l=[97,34,54,23,19,3,78]
print(l)
heapq.heapify(l)
print(l)
heapq.heappush(l,0)
print(l)
print(heapq.heappop(l))
print(l)
heapq.heapreplace(l,6)
print(l)
```

```python
## Graphs in Python

class graph:
    
    def __init__(self,graph):
        if graph is None:
            graph = {}
        self.graph = graph
        
        
    def getVertices(self):
        return list(self.graph.keys())
    
    def getEdges(self):
        result = []
        for vertex,edges in self.graph.items():
            result.extend([{vertex,edge} for edge in edges if {vertex,edge} not in result])

        return result
    
    def insertVertice(self,vertice):
        if vertice not in self.graph:
            self.graph[vertice]=[]
        else:
            print("Vertice already exists!")
            
    def insertEdge(self,edge):
        v1,v2 = tuple(edge)
        if v2 not in self.graph[v1] and v1 not in self.graph[v2]:
            self.graph[v1]=v2
            self.graph[v2]=v1
        else:
            print("Edge already exists!")
    
graph_elements = {'a':['b','e','d'],
     'b':['a','c'],
     'c':['b','e','d'],
     'd':['c','a'],
     'e':['a','c']
    }

g = graph(graph_elements)
print(g.getVertices())    
print(g.getEdges())
g.insertVertice('b')
g.insertVertice('f')
print(g.getVertices())
print(g.getEdges())
g.insertEdge({'b','f'})
print(g.getEdges())
```
